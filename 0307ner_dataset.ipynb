```python
import random

texts = []
for _ in range(0,10):
    texts.append(f'{random.choice(PER)} 선수의 경기 기록이 궁금해.\n')
    texts.append(f'{random.choice(PER)} 선수의 {random.choice(GAL)} 기록이 어떻게 돼?\n')
    texts.append(f'{random.choice(PER)} 선수의 {random.choice(ASS)} 기록이 어떻게 돼?\n')
    texts.append(f'{random.choice(TMM)}팀의 경기 기록이 궁금해.\n')
    texts.append(f'{random.choice(TMM)}의 기록이 궁금해.\n')
    texts.append(f'{random.choice(TMM)}팀의 경기 {random.choice(PRD)}해줘.\n')
```

```python
texts
```

```python
import re

def tag_entities(text, entity_type, entities):
    for entity in sorted(entities, key=len, reverse=True):
        # 길이가 긴 엔티티부터 태깅
        escaped_entity = re.escape(entity)  # 특수 문자가 포함된 엔티티 처리
        # 엔티티 뒤에 바로 '팀'이나 '의'가 오는 경우를 고려
        pattern = rf"({escaped_entity})(팀|의|팀의)?\b"
        replacement = rf"<\1:{entity_type}>\2"
        text = re.sub(pattern, replacement, text)
    return text


tagged_formatted_texts = []  # 결과를 저장할 리스트 초기화

for text in texts:
    original_text = text.strip()
    ner_text = original_text
    ner_text = tag_entities(ner_text, 'PLY', PLY)
    ner_text = tag_entities(ner_text, 'PER', PER)
    ner_text = tag_entities(ner_text, 'TMM', TMM)
    ner_text = tag_entities(ner_text, 'GAL', GAL)
    ner_text = tag_entities(ner_text, 'ASS', ASS)
    ner_text = tag_entities(ner_text, 'RST', RST)
    ner_text = tag_entities(ner_text, 'PRD', PRD)
    ner_text = tag_entities(ner_text, 'STM', STM)

    formatted_text = f"{original_text}\u241E{ner_text}"
    tagged_formatted_texts.append(formatted_text)

tagged_formatted_texts_joined = "\n".join(tagged_formatted_texts)
print(tagged_formatted_texts_joined)
```
